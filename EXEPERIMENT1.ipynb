A Python program to display Operating System name, version, and architecture using the platform library.


```python
import platform
import sys

def get_os_info():
    try:
        os_name = platform.system()        # e.g., 'Windows', 'Linux'
        os_version = platform.version()    # Detailed OS version
        architecture = platform.machine()  # e.g., 'x86_64', 'AMD64'

        print("Operating System Information:")
        print(f"  Name        : {os_name}")
        print(f"  Version     : {os_version}")
        print(f"  Architecture: {architecture}")

    except Exception as e:
        print(f"Error retrieving OS information: {e}", file=sys.stderr)

if __name__ == "__main__":
    get_os_info()
```
