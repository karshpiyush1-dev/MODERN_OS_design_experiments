```python
def fifo(pages, capacity):
    memory = []
    faults = 0

    for p in pages:
        if p not in memory:
            faults += 1
            if len(memory) == capacity:
                memory.pop(0)
            memory.append(p)
        print(memory)

    print("FIFO Page Faults:", faults)


def lru(pages, capacity):
    memory = []
    faults = 0

    for p in pages:
        if p not in memory:
            faults += 1
            if len(memory) == capacity:
                memory.pop(0)
        else:
            memory.remove(p)
        memory.append(p)
        print(memory)

    print("LRU Page Faults:", faults)


pages = [7, 0, 1, 2, 0, 3, 0, 4]
capacity = 3

fifo(pages, capacity)
lru(pages, capacity)
```
