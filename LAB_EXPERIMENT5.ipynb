```python
def is_safe(processes, avail, maxm, allot):
    need = []
    for i in range(len(processes)):
        need.append([maxm[i][j] - allot[i][j] for j in range(len(avail))])

    finish = [False] * len(processes)
    safe_seq = []
    work = avail.copy()

    while len(safe_seq) < len(processes):
        found = False
        for i in range(len(processes)):
            if not finish[i] and all(need[i][j] <= work[j] for j in range(len(avail))):
                for j in range(len(avail)):
                    work[j] += allot[i][j]
                safe_seq.append(processes[i])
                finish[i] = True
                found = True

        if not found:
            print("System is not in safe state")
            return

    print("System is in safe state")
    print("Safe Sequence:", safe_seq)


processes = [0, 1, 2]
avail = [3, 3, 2]

maxm = [[7, 5, 3],
        [3, 2, 2],
        [9, 0, 2]]

allot = [[0, 1, 0],
         [2, 0, 0],
         [3, 0, 2]]

is_safe(processes, avail, maxm, allot)
```
