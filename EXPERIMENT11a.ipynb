Write a Python program to simulate memory allocation (First Fit, Best Fit, Worst Fit).
-->> FIRST FIT

```python
block =[100,500,200,300,600]
processes =[212,417,112,426,99]
for p in processes:
    ALLOCATED = False
    for i in range(len(block)):
        if block [i]>=p:
            print(f"process {p}-> Block {i} allocated")
            block[i]-=p
            ALLOCATED =True
            break
    if not ALLOCATED:
        print(f"process{p}-> not Allocated")
```
