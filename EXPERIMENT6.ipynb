Implementing The First Come First Serve (FCFS) CPU scheduling in Python.

```python
n = int(input("Enter the number of processes: "))
process_burst_time = []
for i in range(n):
    process_burst_time.append(int(input(f"Enter burst time for process {i+1}: ")))
process_wait_time = [0] * n
process_turnaround_time=[0] * n
for i in range(1, n):
    process_wait_time[i] = process_wait_time[i - 1] + process_burst_time[i - 1]
for i in range(n):
    process_turnaround_time[i] = process_wait_time[i] + process_burst_time[i]
print("Process\tBurst Time\tWaiting Time\tTurnaround Time")
for i in range(n):
    print(f"P{i+1}\t{process_burst_time[i]}\t\t{process_wait_time[i]}\t\t{process_turnaround_time[i]}")
avg_wait_time = sum(process_wait_time) / n
avg_turnaround_time = sum(process_turnaround_time) / n
print(f"Average Waiting Time: {avg_wait_time}")
print(f"Average Turnaround Time: {avg_turnaround_time}")
```
